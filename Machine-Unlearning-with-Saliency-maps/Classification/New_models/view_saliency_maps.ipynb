```python
import torch
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the dictionary from the .pt file
state_dict = torch.load('/home/ubuntu/IDL-Project-Machine-Unlearning/Machine-Unlearning-with-Saliency-maps/Classification/saliency_map_path/with_0.5.pt')

# Inspect the dictionary keys
print("Keys in the dictionary:", state_dict.keys())
```

```python
# Access the saliency maps using the appropriate key
# Replace 'saliency_maps_key' with the actual key in your dictionary
# For example, if the saliency maps are stored under 'layer1.0.conv1.weight'
saliency_maps = state_dict['fc.weight']

# Check the type and shape of the saliency maps
print(type(saliency_maps))
print(saliency_maps.shape)
```

```python
# Convert the saliency maps to a NumPy array if necessary
if isinstance(saliency_maps, torch.Tensor):
    saliency_maps = saliency_maps.cpu().numpy()
```

```python
# Visualize the weights
# Assuming fc_weights is a 2D tensor with shape (out_features, in_features)

# For simplicity, let's visualize the weights as a heatmap
plt.imshow(saliency_maps, cmap='hot', aspect='auto')
plt.colorbar()
plt.title('Fully Connected Layer Weights')
plt.xlabel('Input Features')
plt.ylabel('Output Features')
plt.show()
```
